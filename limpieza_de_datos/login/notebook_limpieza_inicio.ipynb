```python
import pandas as pd

# Link en el AWS: Propiedades > URL del objeto
url1 = "https://desafiogrupo1.s3.us-east-1.amazonaws.com/rba_dataset.xlsx"
df1 = pd.read_excel(url1)
df1.head()
```

```python
num_no_nan = df1["Round-Trip Time [ms]"].count()

print("Cantidad de valores no NaN:", num_no_nan)
```

```python
df1['Login Timestamp'] = pd.to_datetime(df1['Login Timestamp'])

# Ordenar de la más antigua a la mas reciente
df1_ordenado = df1.sort_values(by='Login Timestamp', ascending=False)

# Mostrar las primeras filas
df1_ordenado.head(40)
```

```python
# Filtrar por combinaciones
df_rojo = df1_ordenado[(df1_ordenado['Login Successful'] == True) & 
                       (df1_ordenado['Is Attack IP'] == True) & 
                       (df1_ordenado['Is Account Takeover'] == True)]

df_naranja = df1_ordenado[(df1_ordenado['Login Successful'] == True) & 
                          (df1_ordenado['Is Attack IP'] == True) & 
                          (df1_ordenado['Is Account Takeover'] == False)]

df_amarillo = df1_ordenado[(df1_ordenado['Login Successful'] == False) & 
                           (df1_ordenado['Is Attack IP'] == True) & 
                           (df1_ordenado['Is Account Takeover'] == False)]

df_verde = df1_ordenado[(df1_ordenado['Login Successful'] == True) & 
                        (df1_ordenado['Is Attack IP'] == False) & 
                        (df1_ordenado['Is Account Takeover'] == False)]

# Mostrar la cantidad de filas en cada df
print("DF ROJO:", len(df_rojo))
print("DF NARANJA:", len(df_naranja))
print("DF AMARILLO:", len(df_amarillo))
print("DF VERDE:", len(df_verde))
```

```python
# Tomar el sample de cada df
sample_rojo = df_rojo.sample(n=7, random_state=42)
sample_naranja = df_naranja.sample(n=15, random_state=42)
sample_amarillo = df_amarillo.sample(n=30, random_state=42)
sample_verde = df_verde.sample(n=98, random_state=42)

# Unir todos los samples en un solo df
df_sample_total = pd.concat([sample_rojo, sample_naranja, sample_amarillo, sample_verde])

# Opcional: reordenar por timestamp si quieres
df_sample_total = df_sample_total.sort_values(by='Login Timestamp', ascending=False)

# Mostrar la cantidad de registros para verificar
print(df_sample_total['Login Successful'].value_counts())
print(len(df_sample_total))
```

```python
# Dividir cada color en 80% y 20%
df1_rojo = sample_rojo.sample(n=6, random_state=42)
df2_rojo = sample_rojo.drop(df1_rojo.index)

df1_naranja = sample_naranja.sample(n=12, random_state=42)
df2_naranja = sample_naranja.drop(df1_naranja.index)

df1_amarillo = sample_amarillo.sample(n=24, random_state=42)
df2_amarillo = sample_amarillo.drop(df1_amarillo.index)

df1_verde = sample_verde.sample(n=78, random_state=42)
df2_verde = sample_verde.drop(df1_verde.index)

# Unir por DataFrame
df1_final = pd.concat([df1_rojo, df1_naranja, df1_amarillo, df1_verde])
df2_final = pd.concat([df2_rojo, df2_naranja, df2_amarillo, df2_verde])

# Opcional: reordenar por timestamp si quieres
df1_final = df1_final.sort_values(by='Login Timestamp', ascending=False)
df2_final = df2_final.sort_values(by='Login Timestamp', ascending=False)

# Verificar cantidad de registros por color
print("DF1 counts:")
print(df1_final.groupby(['Login Successful', 'Is Attack IP', 'Is Account Takeover']).size())
print("Total DF1:", len(df1_final))

print("\nDF2 counts:")
print(df2_final.groupby(['Login Successful', 'Is Attack IP', 'Is Account Takeover']).size())
print("Total DF2:", len(df2_final))
```

```python
df1_final.to_excel("df1_final.xlsx", index=False)
df2_final.to_excel("df2_final.xlsx", index=False)
```

```python
df1_final.to_csv("df1_final.csv", index=False)
df2_final.to_csv("df2_final.csv", index=False)
```
